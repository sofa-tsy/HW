```python
#*Создайте рандомный вектор и выполните нормализацию столбцов 
#(из каждого столбца вычесть среднее этого столбца, из каждого столбца вычесть sd этого столбца)
g = np.random.randint(0,50,10)
print(g)
print(g.mean())
print(g-g.mean())
print(g.std())
print(g-g.std())
```
